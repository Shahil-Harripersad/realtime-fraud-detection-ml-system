```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('E:/Program Files (x86)/GitHub/datasets/creditcard.csv')
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head()
```

# 1. Class Imbalance

```python
class_counts = df["Class"].value_counts()

print(class_counts)

fraud_ratio = class_counts[1] / len(df)
print(f"Fraud ratio: {fraud_ratio:.4%}")
```

# 2. Missing values

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(f"{col} has {df[col].isnull().sum()} missing values.")

if df.isnull().sum().sum() == 0:
    print("No missing values in the dataset.")
```

# 3. Summary Statistics

```python
df.describe()
```

# 4. Distributions

```python
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

fig, ax1 = plt.subplots(figsize=(8,4))

# Normal transactions
sns.histplot(
    normal["Amount"],
    bins=100,
    color="blue",
    label="Normal",
    log_scale=True,
    ax=ax1
)

ax1.set_ylabel("Normal Count", color="blue")

# Fraud transactions
ax2 = ax1.twinx()

sns.histplot(
    fraud["Amount"],
    bins=100,
    color="red",
    label="Fraud",
    log_scale=True,
    ax=ax2
)

ax2.set_ylabel("Fraud Count", color="red")

ax1.set_xlabel("Amount")
ax1.set_title("Transaction Amount Distribution")

plt.show()
```

```python
# Note: dataset spans 2 days which explains the two peaks in the time distribution
plt.figure(figsize=(10,4))
sns.histplot(df["Time"], bins=100)
plt.title("Transaction Time Distribution")
plt.show()
```

# 5. Correlation analysis

```python
corr = df.corr()

class_corr = corr['Class'].abs().sort_values(ascending=False)
top_features = class_corr[1:6]
print("Top 5 features most correlated with Class:")
for feature, corr_val in top_features.items():
    print(f"{feature}: {corr_val:.4f}")

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
features = ["V10","V12","V14","V17"] # top 4 features most correlated with Class

for f in features:
    plt.figure()
    sns.kdeplot(normal[f], label="Normal")
    sns.kdeplot(fraud[f], label="Fraud")
    plt.title(f"{f} distribution")
    plt.legend()
```
